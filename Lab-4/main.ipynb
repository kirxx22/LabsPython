```python
from pprint import pprint as pp
import pandas as pd
import numpy as np
```

```python
items = {
    'r': {"size": 3, "score": 25},
    'p': {"size": 2, "score": 15},
    'a': {"size": 2, "score": 15},
    'm': {"size": 2, "score": 20},
    'i': {"size": 1, "score": 5},
    'k': {"size": 1, "score": 15},
    'x': {"size": 3, "score": 20},
    't': {"size": 1, "score": 25},
    'f': {"size": 1, "score": 15},
    'd': {"size": 1, "score": 10},
    's': {"size": 2, "score": 20},
    'c': {"size": 2, "score": 20}
}
```

```python
score = 15 # Начальный счет
for key in items.keys():
    score -= items[key]['score']

y, x = len(items), 10

# inventory = np.zeros((len(items), 10), dtype=np.int64)
inventory = np.full((len(items), 10), score)
inventory_items = [[[] for j in range(10)] for i in range(len(items))]

data = pd.DataFrame(inventory, index=items.keys(), columns=range(x))
data
```

```python
# заполняем для первого предмета
for j in range(x):
    # if data.iloc[0, j]
    if j >= items['r']['size']:
        data.iloc[0, j] = (items['r']['score']) * 2 + score
        inventory_items[0][j].append('r')

data
```

```python

for i, key in zip(range(1, y), list(items.keys())[1:]):
    for j in range(x):
        if j - items[key]['size'] > -1:
            data.iloc[i, j] = max(data.iloc[i - 1, j], items[key]['score'] * 2 + data.iloc[i - 1, j - items[key]['size']])
            if items[key]['score'] * 2 + data.iloc[i - 1, j - items[key]['size']] > data.iloc[i - 1, j]:
                inventory_items[i][j] = inventory_items[i - 1][j - items[key]['size']] + [key]
            else:
                inventory_items[i][j] = inventory_items[i - 1][j]

        else:
            data.iloc[i, j] = data.iloc[i - 1, j]
            inventory_items[i][j] = inventory_items[i - 1][j]

data
```

```python
result_inventory = [[key] for key in inventory_items[-1][-1] for j in range(items[key]['size'])]
result_inventory = np.array(result_inventory).reshape(3, 3)
result_inventory
```

```python
def print_inventory(inventory: np.ndarray):
    for row in inventory:
        for item in row:
            print([item], end=' ')
        print()
        
print_inventory(result_inventory)
print(f'Максимальное количество очков: {data.iloc[-1, -1]}')
```
