````python
data = pd.read_csv('books-en.csv', delimiter=';', encoding='cp1252')
data.info()
````

````python
data.Downloads = data.Downloads.apply(int)
data.Price = data.Price.apply(lambda x: float(x.replace(',', '.')))
data.info()
````

### Задание № 1

Вывести количество записей, у которых в поле Название строка длиннее 30 символов.

````python
count = len([i for i in data['Book-Title'].values if len(i) > 30])
count
````

### Задание № 2

Реализовать поиск книги по автору, использовать ограничение на выдачу в зависимости от варианта.

**Ограничение**: _До 200 рублей_

````python
author = input()
response = data[(data.Price < 200) & (data['Book-Author'] == author)]
response.head()
````

### Задание № 3

Реализовать генератор библиографических ссылок вида ```<автор>. <название> - <год>``` для 20 записей. Записи выбрать произвольно. Список сохраняется как отдельный файл текстового формата с нумерацией строк.


````python
pattern = '{}. {} - {}\n'
f = open('result.txt', 'w')
for _ in range(20):
    row = data.loc[_]
    f.write(pattern.format(row['Book-Author'], row['Book-Title'], row['Year-Of-Publication']))
f.close()
````
